```python
median_suicide_count = data['SuicideCount'].median()
data['SuicideCount'].fillna(median_suicide_count,inplace = True)
```

```python
data.drop(columns=['CauseSpecificDeathPercentage'],inplace = True)
```

```python
median_death_rate = data['DeathRatePer100K'].median()
data['DeathRatePer100K'].fillna(median_death_rate,inplace = True)
```

```python
population_related = ['Population','GDP','GDPPerCapita','GrossNationalIncome','GNIPerCapita','InflationRate','EmploymentPopulationRatio']
for i in population_related:
    median_value = data[i].median()
    data[i].fillna(median_value,inplace=True)
```

### Suicide Rate over Time

```python
suicide_by_year = data.groupby('Year')['SuicideCount'].sum()
plt.figure(figsize=(10,6))
sns.lineplot(x=suicide_by_year.index,y=suicide_by_year.values,marker='o')
plt.title('Trend of Suicide Rates Over Time (Year)')
plt.xlabel('Year')
plt.ylabel('Total Suicide Count')
plt.grid(True)
plt.show()
```

### Regional Analysis of Suicide Rates 

```python
suicide_region = data.groupby('RegionName')['SuicideCount'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=suicide_region.index,y=suicide_region.values)
plt.title('Suicide Rates Across Regions')
plt.xlabel('Region')
plt.ylabel('Total Suicide Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Categorical Analysis

```python
suicide_by_age = data.groupby('AgeGroup')['SuicideCount'].sum().sort_values(ascending=False)
suicide_by_sex = data.groupby('Sex')['SuicideCount'].sum()
suicide_by_generation = data.groupby('Generation')['SuicideCount'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
# Plot suicide rates by sex
plt.subplot(1, 3, 1)
sns.barplot(x=suicide_by_sex.index, y=suicide_by_sex.values)
plt.title('Suicide Rates by Sex')
plt.xlabel('Sex')
plt.ylabel('Total Suicide Count')

# Plot suicide rates by age group
plt.subplot(1, 3, 2)
sns.barplot(x=suicide_by_age.index, y=suicide_by_age.values)
plt.title('Suicide Rates by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Suicide Count')
plt.xticks(rotation=90)

# Plot suicide rates by generation
plt.subplot(1, 3, 3)
sns.barplot(x=suicide_by_generation.index, y=suicide_by_generation.values)
plt.title('Suicide Rates by Generation')
plt.xlabel('Generation')
plt.ylabel('Total Suicide Count')
plt.xticks(rotation=90)
#plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### Correlation Analysis

```python
correlation_variables = ['SuicideCount', 'GDP', 'GDPPerCapita', 'Population', 'InflationRate', 'EmploymentPopulationRatio']
correlation_matrix = data[correlation_variables].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='EmploymentPopulationRatio', y='SuicideCount')
plt.title('Scatter Plot: Suicide Count vs. EmploymentPopulationRatio')
plt.xlabel('EmploymentPopulationRatio')
plt.ylabel('Suicide Count')
plt.show()
```
